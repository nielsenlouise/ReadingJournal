### Section 4.3 exercises
Work through exercises 1-5 in Section 4.3, creating the `square`, `polygon`, `circle`, and `arc` functions. You can reference Allen's solutions, but please try the problems on your own first.

````python
from math import pi

# make a world, turtle a world
world = TurtleWorld()
thing = Turtle()

def square(t):
    """t is the name of the turtle. it will draw a square.
    """
# thing makes a square
    for i in range(4):
        fd(t, 100)
        lt(t)

def square(t, length):
    """t is the name of the square-drawing turtle.
    length is the length of each side.
    """
    for i in range(4):
        fd(t,length)
        lt(t)
        
def polygon(t, length, n):
    """t is the name of the polygon-drawing turtle.
    length is the length of each side.
    n is the number of sides
    """
    for i in range(n):
        fd(t,length)
        lt(t,360./n)
        
def circle(t,r):
    """t is the name of the circle-drawing turtle.
    r is the radius of the circle to be drawn.
    """
    circumf = 2 * pi * r
    n = int(circumf/3)
    length = circumf/n
    polygon(t,length,n)

def arc(t,r,angle):
    """t is the name of the arc-drawing turtle.
    r is the radius of the arc to be drawn.
    angle is the angle to be arced (in degrees).
    """
    circumf = 2 * pi * r
    frac = angle/360.
    arc_length = circumf * frac
    n = int(arc_length/3) + 1
    side_length = arc_length/n
    side_angle = float(angle)/n
    for i in range(n):
        fd(t,side_length)
        lt(t,side_angle)

thing.delay = .01
arc(thing,50,270)
wait_for_user()
````
